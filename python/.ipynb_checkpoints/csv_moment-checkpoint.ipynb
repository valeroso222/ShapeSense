```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
#csvを読み込む
def csv_to_array(file_name):
    csv_input = pd.read_csv("Data/"+file_name+".csv")
    data = np.round(np.array(csv_input.values)*50)*2
    data_max_value = 0
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            if math.isnan(data[i][j]):
                data[i][j] = 0
            if data[i][j] > data_max_value:
                data_max_value = data[i][j]
    print(data_max_value)
    return (data, data_max_value)

def make_plot_data(data, data_max_value):
    plot_data = np.array([np.zeros((data.shape[1], int(data_max_value))) for i in range(data.shape[0])])
    for i in range(plot_data.shape[0]):
        for j in range(plot_data.shape[1]):
            for l in range(int((plot_data.shape[2] -data[i][j])/2), int((plot_data.shape[2]+data[i][j])/2)):
                plot_data[i][j][l] = 1
        plot_data[i] = plot_data[i][::-1]
    return plot_data

def conv_to_RGB(data):
    list = [[]for i in range(data.shape[0])]
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            if data[i][j] == 1:
                list[i].append([0.0,0.0,0.0])
            else:
                list[i].append([1.0,1.0,1.0])
    return list

def plot_3(file_name):
    #csvをcsv配列に
    data_ans, data_max_value_ans = csv_to_array("answer")
    data, data_max_value = csv_to_array(file_name)
    data_max_value = max(data_max_value, data_max_value_ans)
    #配列をplot配列に
    plot_data_ans = make_plot_data(data_ans, data_max_value)
    plot_data = make_plot_data(data, data_max_value)
    for i in range(plot_data_ans.shape[0]):
        print(i+1)
        #plot配列[i]をRGB配列に
        plot_data_RGB_ans = np.array(conv_to_RGB(plot_data_ans[i%plot_data_ans.shape[0]]))
        plot_data_RGB_1 = np.array(conv_to_RGB(plot_data[i]))
        plot_data_RGB_2 = np.array(conv_to_RGB(plot_data[i + plot_data_ans.shape[1]]))
        fig = plt.figure()
        ax1 = fig.add_subplot(131)
        ax2 = fig.add_subplot(132)
        ax3 = fig.add_subplot(133)
        ax1.imshow(plot_data_RGB_ans)
        ax2.imshow(plot_data_RGB_1)
        ax3.imshow(plot_data_RGB_2)
        ax1.axis("off")
        ax2.axis("off")
        ax3.axis("off")
        ax1.set_aspect(4)
        ax2.set_aspect(4)
        ax3.set_aspect(4)
        plt.show()

def plot_2(file_name):
    #csvをcsv配列に
    data_ans, data_max_value_ans = csv_to_array("answer")
    data, data_max_value = csv_to_array(file_name)
    data_max_value = max(data_max_value, data_max_value_ans)
    #配列をplot配列に
    plot_data_ans = make_plot_data(data_ans, data_max_value)
    plot_data = make_plot_data(data, data_max_value)
    for i in range(plot_data_ans.shape[0]):
        print(i+1)
        #plot配列[i]をRGB配列に
        plot_data_RGB_ans = np.array(conv_to_RGB(plot_data_ans[i%plot_data_ans.shape[0]]))
        plot_data_RGB = np.array(conv_to_RGB(plot_data[i]))
        fig = plt.figure()
        ax1 = fig.add_subplot(121)
        ax2 = fig.add_subplot(122)
        ax1.imshow(plot_data_RGB_ans)
        ax2.imshow(plot_data_RGB)
        ax1.axis("off")
        ax2.axis("off")
        ax1.set_aspect(4)
        ax2.set_aspect(4)
        plt.show()

def moment(file_name):
    f = open("Data/moment.csv", "a")
    csvWriter = csv.writer(f, lineterminator="\n")
    data_ans, data_max_value_ans = csv_to_array(file_name)
    plot_data_ans = make_plot_data(data_ans, data_max_value_ans)
    inertia_data_ans = np.array([plot_data_ans[i][::-1] for i in range(plot_data_ans.shape[0])])
    I_u = np.zeros(inertia_data_ans.shape[0])
    I_v = np.zeros(inertia_data_ans.shape[0])
    for l in range(inertia_data_ans.shape[0]):
        for i in range(inertia_data_ans.shape[1]):
            for j in range(inertia_data_ans.shape[2]):
                if inertia_data_ans[l][i][j] == 1:
                    I_u[l] += (4*i+0.5)**2
                    I_u[l] += (4*i+1.5)**2
                    I_u[l] += (4*i+2.5)**2
                    I_u[l] += (4*i+3.5)**2
                    I_v[l] += (abs(j-(data_max_value_ans-1)/2))**2*4
    #print(I_u)
    #print(I_v)
    csvWriter.writerow(I_u)
    csvWriter.writerow(I_v)
    f.close()
```

```python
moment("answer")
moment("kameda")
moment("yamamoto")
moment("yoshino")
moment("kobatake")
moment("ichimura")
moment("morita")
moment("shimoyama")
moment("matsunaga")
moment("michizono")
moment("nakagawa")
```
