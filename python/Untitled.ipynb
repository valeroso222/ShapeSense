```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4
 
fig, axs = plt.subplots(3, 2, figsize=(6, 4))
 
# 左上
axs[0, 0].plot(x, y1)
 
# 右上
axs[0, 1].plot(x, y2)
 
# 左下
axs[1, 0].plot(x, y3)
 
# 右下
axs[1, 1].plot(x, y4)
 
plt.show()
```
