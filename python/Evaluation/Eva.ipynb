```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
A1=[1,3,1,1,4,4,1,1,1,4,1,1]
A2=[2,4,3,4,3,2,1,2,2,2,3,2]
A3=[1,3,4,2,4,5,4,3,3,3,2,3]
A4=[4,4,3,2,1,4,2,1,4,2,1,4]
A5=[5,5,2,5,5,5,4,5,5,5,5,5]
percieved=A1+A2+A3+A4+A5
true=[1 for i in range(12)]+[2 for i in range(12)]+[3 for i in range(12)]+[4 for i in range(12)]+[5 for i in range(12)]
labels_s=['S1','S2','S3','S4','S5']
labels_v=['V1','V2','V3','V4','V5']
cm=confusion_matrix(true,percieved)
print('accuracy',accuracy_score(true,percieved))
print('precision',precision_score(true,percieved,average=None))
print('recall',recall_score(true,percieved,average=None))
print('f1',f1_score(true,percieved,average=None))
df=pd.DataFrame(cm,index=labels_s,columns=labels_v)/0.12
plt.figure(figsize=(2,1.5))
sns.heatmap(df,annot=True,cmap="Reds",vmax=100,cbar=False)
#plt.title("Confusion Matrix")
plt.xlabel("Virtual Shape")
plt.ylabel("Actual Shape")
plt.savefig("image/ConfusionMatrixSS1.png",dpi=300,bbox_inches="tight")
plt.show()
```
